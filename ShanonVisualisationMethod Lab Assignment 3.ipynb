```python
from collections import defaultdict
import random
```

```python
def readFile(fileName):
    with open(fileName) as f:
        return [line.strip().lower() for line in f.readlines()]
```

```python
x = readFile('txt1.txt');
```

### data => defaultDict[(gram[i], gram[i+1], ....)] = frequency

```python
class shannonVisualisationMethod():
    
    def __init__(self, which_gram, data):
        self.which_gram = which_gram;
        self.data = self.transform(which_gram, data);
    
    def transform(self, which_gram, data):
        calculate_frequency = defaultdict(int)
        #print(data)
        for sentence in data:
            list_of_sent = sentence.split();
            list_of_sent.insert(0,'<>')
            list_of_sent.append('</>')
            start = 0;
            for word in range(len(list_of_sent) - which_gram + 1):
                calculate_frequency[tuple(list_of_sent[start: start + which_gram])] += 1
                start += 1
        return calculate_frequency;
    
    def start_forming_sent(self, max_iteration=20):
        curr = tuple()
        while max_iteration:
            max_iteration-=1
            if len(curr) == 0:
                curr = self.select_first_word()
                print(' '.join(curr[1:]),end=' ')
                continue;
            else:
                curr = self.next_(curr);

            if curr[self.which_gram - 1] == '</>':
                print('.',end='');
                break;
            else:        
                print(curr[-1],end=' ')
        print(end='\n')

    def select_first_word(self):
        temp = [[],[]];
        
        for k, v in self.data.items():
            if k[0] == '<>':
                temp[0].append(k)
                temp[1].append(v)
          
        return random.choices(temp[0], weights=temp[1], k=1)[0];
    
    def next_(self, curr):
        temp = [[],[]];
        
        for k, v in self.data.items():
            if k[:-1] == curr[1:]:
                temp[0].append(k)
                temp[1].append(v)
          
        return random.choices(temp[0], weights=temp[1], k=1)[0];
            
```

# Using Bigram Probabilities

```python
bigram_shannon = shannonVisualisationMethod(2,x)
```
